```python
def LU(A):

    '''Do the guassian elemenation'''
    '''in fact, this is also called doolittle's algorithm'''
    
    U=A.copy()
    tol=1e-6
    dim=U.shape[0]
    det=1
    L=np.eye(dim,dtype=float)
    P=L.copy()
    for i in range(1,dim):
        det*=U[i-1,i-1]
        for j in range(i,dim):
            L[j,i-1]=(U[j,i-1]/U[i-1,i-1])
            U[j,:]-=(U[j,i-1]/U[i-1,i-1])*U[i-1,:]
    return U,L,det
```

```python
#input depend on question and user
A=np.zeros((5,5))
for i in range(5):
    for j in range(5):
        if j>i:
            A[i,j]=0
        else:
            A[i,j]=2*(i+1)/(i+j+2)

dim=A.shape[0]
f=np.ones((dim,1),dtype=float) #constant term
```

```python
# the cholesky decomposition, note that cholesky is only valid for symmetric positive def mat.
# it is very common to see L.T!=U if the matrix does not satisfy the cond.
def cholesky(A):
    U,L,det=LU(A)
    dim=U.shape[0]
    D=np.diag(np.sqrt(np.diag(U)))
    for i in range(dim):
        U[i,:]/=np.diag(U)[i]
    L=L@D
    U=D@U
    return L,D,U
```

```python
L,D,U=cholesky(A)
print(L)
```

```python
def RREF(A):
    U=LU(A)[0]
    dim=U.shape[0]
    
    for i in range(dim-1,0,-1):
        for j in range(i):
            U[j,:]-=(U[j,i]/U[i,i])*U[i,:]
            U[j,:]/=U[j,j]
            #print(U)
    U[dim-1,:]/=U[dim-1,dim-1]
    return U
```

```python
# some special matrices, 
# the inputs are the information that we only need when constructing the matrices

def tridiag(dim,l,m,u):
    '''dimension, lower diag, middle diag, upper diag'''
    T=np.zeros((dim,dim))
    T[0,0],T[0,1]=m,u
    T[-1,-1],T[-1,-2]=m,l
    for i in range(1,dim-1):
        T[i,i-1]=l
        T[i,i]=m
        T[i,(i+1)%dim]=u
    return T

def DFT(dim):
    '''the dst matrix is just the negative of 1st quater of this matrix,
    and the inverse of it is the 1st quater of this matrix'''
    
    w=np.exp(-2j*np.pi/dim)
    D=np.ones((dim,dim),dtype=complex)
    for i in range(dim):
        for j in range(dim):
            D[i,j]=w**(i*j)
    return D

def iDFT(dim):
    w=np.exp(2j*np.pi/dim)
    D=np.ones((dim,dim),dtype=complex)
    for i in range(dim):
        for j in range(dim):
            D[i,j]=w**(i*j)
    D/=dim
    return D

def circulant(vector): 
    vec=vector.copy()
    dim=len(vec)
    V=np.zeros((dim,dim))
    inter=tridiag(dim,0,0,1)
    inter[-1,0]=1
    V[0,:]=vec
    for i in range(1,dim):
        vec=vec.dot(inter)
        V[i,:]=vec
    return V
```
